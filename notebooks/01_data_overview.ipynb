## В датасете есть следующие признаки:

- **User ID** — идентификатор пользователя  
- **Group** — принадлежность к контрольной или тестовой группе  
- **Page Views** — количество просмотренных страниц  
- **Time Spent** — время нахождения на сайте (в секундах)  
- **Conversion** — факт совершения целевого действия  
- **Device** — тип устройства пользователя  
- **Location** — регион Великобритании  

Посмотрим статистические характеристики всех признаков

```python
df.describe(include = 'all')
```

Посмотрим уникальные значения категориальных признаков

```python
categorical_columns = ['Group', 'Conversion', 'Device', 'Location']

for col in categorical_columns:
    print(f"{col}:")
    print(df[col].unique())
```

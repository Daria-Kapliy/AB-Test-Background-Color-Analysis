## Результаты анализа коэффициента конверсии

Тестовая группа B демонстрирует более высокий коэффициент конверсии по сравнению с контрольной группой A.

Разница между группами составила:

- группа B: около 14% конверсии.

Тестовый вариант обеспечивает заметный прирост вероятности совершения целевого действия пользователем.

## Результаты статистической проверки

Для проверки статистической значимости различий был использован критерий хи-квадрат.
Полученное значение p-value оказалось ниже уровня значимости 0.05, что позволяет отвергнуть нулевую гипотезу об отсутствии различий между группами.
Дополнительная оценка силы связи с помощью коэффициента Cramer's V показала наличие статистически значимой, но слабой связи между принадлежностью пользователя к группе и фактом совершения конверсии.

## Анализ поведенческих метрик

Для количественных метрик поведения пользователей были использованы критерии Манна–Уитни.

- статистически значимого различия по времени, проведённому на сайте, между группами не обнаружено;
- статистически значимого различия по количеству просмотренных страниц не обнаружено.

Значит изменение цвета фона влияет прежде всего на вероятность совершения конверсии, но не сопровождается выраженным изменением общей пользовательской активности.
